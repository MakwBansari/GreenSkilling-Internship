# **Generalization of the Dataset**

- Generalization is the process of replacing specific values with broader categories to improve privacy and simplify data analysis.

- In this sustainability dataset, the following generalizations can be applied:

  - Entity (Country Name) can be generalized to Country or Region.
  - GDP per Capita can be categorized as Low Income, Middle Income, or High   Income instead of using exact values.
  - CO₂ Emissions can be grouped into Low, Medium, and High Emissions.
  - Renewable Energy Share (%) can be classified as Low, Moderate, and High Renewable Usage.
  - Access to Electricity (%) can be categorized into Poor, Moderate, and Excellent Access.
  - Primary Energy Consumption can be grouped as Low, Medium, and High Consumption.
  - Financial Flows to Developing Countries can be classified into Low, Moderate, and High Investment.
  - Population Density can be grouped into Low, Medium, and High Population Density.
  
# **Purpose of Generalization**
  - Protects sensitive information.
  - Makes data easier to interpret.
  - Helps in data anonymization and privacy preservation.
  - Simplifies comparison between countries and regions.
  - Improves readability for analysis and visualization.

# **Handling Missing Values**

```python
green_df1 = green_df.copy()

# Filling Numerical Missing Values with Mean

numeric_columns = green_df1.select_dtypes(include="number").columns

green_df1[numeric_columns] = green_df1[numeric_columns].fillna(
    green_df1[numeric_columns].mean()
)

# Fillin Categorical Missing Values

green_df1["Entity"] = green_df1["Entity"].fillna("Unknown Country")

print("\nMissing Values After Cleaning")
print(green_df1.isnull().sum())
```

# **Forward Fill and Backward Fill**

```python
forward_df = green_df.ffill()

print("\nForward Filled Dataset")
print(forward_df.head())

backward_df = green_df.bfill()

print("\nBackward Filled Dataset")
print(backward_df.head())
```
