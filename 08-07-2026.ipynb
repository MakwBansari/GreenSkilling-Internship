```python
import pandas as pd

# Read CSV file
data = pd.read_csv("employment_status.csv")

# 1. Display the complete dataset
print("1. Complete Dataset")
print(data)

# 2. Display first 5 rows
print("\n2. First 5 Rows")
print(data.head())

# 3. Display last 5 rows
print("\n3. Last 5 Rows")
print(data.tail())

# 4. Display shape of dataset
print("\n4. Shape")
print(data.shape)

# 5. Display column names
print("\n5. Columns")
print(data.columns)

# 6. Display dataset information
print("\n6. Information")
print(data.info())

# 7. Display data types
print("\n7. Data Types")
print(data.dtypes)

# 8. Check for missing values
print("\n8. Missing Values")
print(data.isnull().sum())

# 9. Display unique employment statuses
print("\n9. Unique Values")
print(data["Employment Status"].unique())

# 10. Count each employment status
print("\n10. Value Counts")
print(data["Employment Status"].value_counts())
```
