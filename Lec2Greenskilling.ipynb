## List with 5 methods

```python
my_list = [10,20,30,40,22,44]

my_list.append(50)
print("Append:", my_list)       #appends

my_list.insert(2, 25)
print("Insert:", my_list)          #inserts these elements

my_list.remove(20)
print("Remove:", my_list)        #removes the elment 20

my_list.pop()
print("Pop:", my_list)           #pops the top most element i.e., works like stack

my_list.reverse()
print("Reverse:", my_list)        #reverses the order of list

print("Indexing In List", my_list[2:10])
print("1)", my_list[4:10:3])   # [4 starting including : 10 stop excluding : 3 Step]
print("2)", my_list[::-3])
print("3)", my_list[:-4])

print(type(my_list))
```

## Dictionary with 5 methods

```python
oxford = {
    "Sad": "Misery",
    "Happiness": "Elation",
    "Anger": "Wrath"
}

print("1)", type(oxford))
print("2)", oxford)

# 1. update()
oxford.update({"Fear": "Dread"})
print("3)", oxford)

# 2. items()
print("4)", oxford.items())

# 3. keys()
print("5)", oxford.keys())

# 4. values()
print("6)", oxford.values())

# 5. pop()
oxford.pop("Anger")
print("7)", oxford)
```

## Sets with 5 methods

```python
languages = {"Python", "Java", "C", "JavaScript"}

print("1) ", type(languages))
print("2) ", languages)

# 1. add()
languages.add("C++")
print("3) ", languages)

# 2. update()
languages.update(["Go", "Swift"])
print("4) ", languages)

# 3. remove()
languages.remove("Java")
print("5) ", languages)

# 4. discard()
languages.discard("Ruby")    # No error even if element doesn't exist
print("6) ", languages)

# 5. pop()
removed = languages.pop()
print("    Removed Element:", removed)
print("7) ", languages)
```

## If-elif-else loop example  

```python
score = 78

if score >= 90:
    print("Grade A")
elif score >= 75:
    print("Grade B")
elif score >= 50:
    print("Grade C")
else:
    print("Fail")
```

## Nested if-else loop

```python
marks = int(input("Enter your marks: "))
attendance = int(input("Enter attendance percentage: "))

if marks >= 35:
    if attendance >= 75:
        print("Congratulations! You have passed the exam.")
    else:
        print("You passed the exam but attendance is below 75%.")
else:
    if attendance >= 75:
        print("Good attendance, but you failed the exam.")
    else:
        print("You failed the exam and attendance is also low.")
```

## Break Statement

```python
products = ["Milk", "Bread", "Butter", "Juice", "Rice", "Sugar"]

search = input("Enter product to search: ")

for item in products:
    if item == search:
        print(search, "is available in the shop.")
        break
    print("Checking:", item)
else:
    print(search, "is not available.")
```

## Continue

```python
numbers = [12, -5, 18, -2, 25, 0, 31, -9]

print("Positive Numbers:")

for num in numbers:
    if num < 0:
        continue
    print(num)
```

# For Loop

```python
print("------ FOR LOOP ------")

students = {
    "Aaron Warner": 89,
    "Julliette Ferrars": 76,
    "Kenji Kishimoto": 92,
    "Adam Kent": 84,
    "James Kent": 68
}

print("Student Marks")

for name, marks in students.items():
    print(name, ":", marks)

print("\nStudents who scored above 80:")

for name, marks in students.items():
    if marks > 80:
        print(name)
```

# While Loop

```python
balance = 10000

while balance > 0:
    amount = input("Enter withdrawal amount (0 or stop): ")

    if amount.lower() == "stop" or amount == "0":
        print("Transaction cancelled.")
        break

    amount = int(amount)

    if amount <= balance:
        balance = balance - amount
        print("Remaining Balance =", balance)
    else:
        print("Insufficient Balance.")

print("Thank you for using the ATM.")
```
